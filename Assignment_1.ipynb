Q#4: Perform imputation on column ”normalized-losses” and replace missing values with
mean of column

```python
#Here i did 2 things :
#1. First Converted the 'normalized-losses' to numeric type first
#2. Then i calculated mean but first filled the NaN values with mean of it.
test['normalized-losses'] = pd.to_numeric(test['normalized-losses'], errors='coerce')
test['normalized-losses'].fillna(test['normalized-losses'].mean())
```

Q#5: Drop null values (not full column) from price column

```python
#Here i dropped the rows where the 'price' column has NaN values ,saved them in a variable and printed it out.
#i didn't drop the whole column just drop the rows where price is NaN
test = test.dropna(subset=['price'])
print(test)
```

```python
#Here i plot the graph to check the outliers first if any & I have used 2 columns first 'symboling' and the second one 'normalized-losses'
#Here's the distplot for 'normalized-losses'
sns.distplot(test['normalized-losses'])
```

```python
#to remove the outliers from the 'normalized-losses' i limit the data of 'normalized-losses' that if the 'normalized-losses' data is greater than or equals to 50 & data is less than or equals to 250 and then plot it to check if the outliers is removed well-fully.
test = test[(test['normalized-losses'] >= 50) & (test['normalized-losses'] <= 250)]
sns.distplot(test['normalized-losses'])
```

```python
#Here's the distplot for 'symboling'
sns.distplot(test['symboling'])
```

```python
#to remove the outliers from the 'symboling' i limit the data of 'symboling' that if the 'symboling' (data is greater than or equals to 2.7) & (data is less than or equals to 5)
#and then plot it to check if the outliers is removed well-fully.
tests = test[(test['symboling'] <= 5 )& (test['symboling'] >= 2.7)]
sns.distplot(tests['symboling'])
```

If you have a dataset with 100 samples and set test_size=0.2:
*   X_train and y_train will contain 80 samples (80% of the data for training).
*   X_test and y_test will contain 20 samples (20% of the data for testing).
*   test_size=0.2: Specifies the proportion of the data to include in the test set. In this case, 20% of the data will go into the test set, and the remaining 80% will go into the training set. If you wanted a different split, say 30% for testing, you'd set test_size=0.3.

**If you do not specify test_size, the default split is 75% training and 25% testing.**
